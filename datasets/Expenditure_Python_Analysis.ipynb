```python
%matplotlib inline
import pandas as pd
from pandas import Series, DataFrame
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
```

```python
df = pd.read_csv('Public_Expenditure_Services.csv')
df
```

```python
df = df.set_index('Unnamed: 0').T.rename_axis('Date').reset_index()
df
```

```python
print(df)
```

```python
total_exp = DataFrame(df, columns = ["Date", "1. General public services", "Total Managed Expenditure"])
print(total_exp)
```

```python
total_exp['Growth Rate'] = total_exp['Total Managed Expenditure'].pct_change().round(4)*100
print (total_exp)
```

```python
total_exp_1 = total_exp[total_exp['Date'].isin(["2001", "2002", "2003", "2004","2005","2006","2007","2008"])]
total_exp_1 = DataFrame(total_exp_1, columns = ["Total Managed Expenditure"])
total_exp_01_08 = total_exp_1.sum()
print(total_exp_01_08)
```

```python
total_exp_2 = total_exp[total_exp['Date'].isin(["2009", "2010", "2011", "2012","2013","2014","2015","2016"])]
total_exp_2 = DataFrame(total_exp_2, columns = ["Total Managed Expenditure"])
total_exp_09_16 = total_exp_2.sum()
print(total_exp_09_16)
```
